```python
import numpy as np 
import getdist
%matplotlib inline
```

```python
file_path='/exports/home/luzhiyu/data/Cosmology_packages/montepython/data/DR12_max_kmin01kmax15/'
```

```python
z = np.array([], 'float64')
aperp = np.array([], 'float64')
apara = np.array([], 'float64')
mslope = np.array([], 'float64')
fAmp = np.array([], 'float64')
# read redshifts and data points
with open(file_path+'results_apara_aperp_mslope_fAmp.txt', 'r') as filein:
    for i, line in enumerate(filein):
        print(line)
        if line.strip() and line.find('#') == -1:
            this_line = line.split()
            # load redshifts and alpha_parallel
            if this_line[1] == 'apara':
                z = np.append(z, float(this_line[0]))
                apara = np.append(apara, float(this_line[2]))
            # load alpha perpendicular
            elif this_line[1] == 'aperp':
                aperp = np.append(aperp, float(this_line[2]))
            # load shape tilt
            elif this_line[1] == 'mslope':
                mslope = np.append(mslope, float(this_line[2]))
            # load f * sigma8
            elif this_line[1] == 'fAmp':
                fAmp = np.append(fAmp, float(this_line[2]))
```

```python
mu=np.array([apara[0],aperp[0],mslope[0],fAmp[0]])
cov=np.loadtxt(file_path+'covtot_apara_aperp_mslope_fAmp.txt')[:4,:4]
```

```python
def multi_gauss(xvec,muvec,cov):
    xvec=np.array(xvec)
    muvec=np.array(muvec)
    cov=np.array(cov)
    n=len(muvec)
    
    xvec=xvec.reshape(-1,1)
    muvec=muvec.reshape(-1,1)
    det=np.linalg.det(cov)
    inv=np.linalg.inv(cov)
    return 1./(2*np.pi)**(n/2.)/np.sqrt(det)*np.exp(-0.5*(xvec-muvec).T@inv@(xvec-muvec))
multi_gauss([.5,2.5],[1,2],[[2,0.5],[0.5,1]])
```

```python
from scipy.stats import multivariate_normal
sample=multivariate_normal.rvs(mean=mu,cov=cov,size=100)
```

```python
from getdist.gaussian_mixtures import GaussianND
from getdist.gaussian_mixtures import MixtureND
from getdist import plots
gauss=GaussianND(mu, cov)

mix=MixtureND([mu],[cov])
g = plots.get_subplot_plotter()
g.triangle_plot([gauss,mix],["param1","param2","param3","param4"],filled=True)
```

```python
new=mix.marginalizedMixture(params=["param1","param2"])
g = plots.get_subplot_plotter()
g.triangle_plot([new],filled=True)
```

```python
def marginalized_gaussian(mean,cov,marg_index):
    new_mean = np.delete(mean,marg_index)
    new_cov = np.delete(cov,marg_index,axis=0)
    new_cov = np.delete(new_cov,marg_index,axis=1)
    return new_mean,new_cov
new_mean,new_cov=marginalized_gaussian(mu,cov,marg_index=[1,2])
new_mix=MixtureND([new_mean],[new_cov],names=["param1","param4"])

new=mix.marginalizedMixture(params=["param1","param4"])
g = plots.get_subplot_plotter()
g.triangle_plot([new,new_mix],filled=True)
```
