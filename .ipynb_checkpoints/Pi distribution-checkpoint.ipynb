```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 计算准确圆周率的马青公式
import time

def cal_pi(n=100):
    start_time = time.time()   # 计算pi计时起点
    w = n + 10                 # 考虑精度，位数增加 k，结果减少 k 位
    b = 10 ** w                # 用整数计算pi  如3.14 * 10^2 = 314
    x1 = b * 4// 5             # 第一项前半部分
    x2 = b // -239             # 第一项后半部分
    sum = x1 + x2              # 第一项的值
    n *= 2
    for i in range(3, n, 2):
        x1 //= -25
        x2 //= -239*239
        x = (x1+x2)//i
        sum += x
    mpi = sum * 4
    mpi //= 10**10
    end_time = time.time()
    run_time = str(end_time - start_time)
    mpi_str = str(mpi)
    mpi_str = mpi_str[:1] + '.' + mpi_str[1:]
    # f = open('mpi.txt', 'w')
    # f.write(mpi_str)
    # f.close()
    return mpi_str

pi=cal_pi(1000)
pi
```

```python
num=3000
pi=cal_pi(num)
number=np.zeros((num,10))
for j in range(num):
    pinew=pi[:-j]
    for i in range(10):
        number[j,i]= pinew.count(str(i))
```

```python
x=np.arange(num)
for i in range(0,10):
    plt.plot(number[:1:-1,i],label=i)
    
plt.legend()
```
