```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# 定义微分方程组的模型函数
def model(y, t):
    x, z = y
    dxdt = -2 * x
    dydt = 3 * x - z
    return [dxdt, dydt]

# 定义初始条件和时间范围
y0 = [1.0, 0.0]  # 初始条件，例如 x(0) = 1.0，y(0) = 0.0
t = np.linspace(0, 5, 101)  # 时间范围，从 0 到 5，包含 101 个时间点

# 使用 odeint 求解微分方程组
solution = odeint(model, y0, t)

# 提取解
x_solution = solution[:, 0]  # 提取 x(t) 的解
y_solution = solution[:, 1]  # 提取 y(t) 的解

# 绘制解的图像
plt.figure(figsize=(8, 6))
plt.plot(t, x_solution, label='x(t)')
plt.plot(t, y_solution, label='y(t)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()
```
